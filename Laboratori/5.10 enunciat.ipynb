### 5.10 Comptador d'inversions en una llista

Volem ordenar una llista A de nombres enters en ordre creixent, es a dir,
$A[i] < A[j]$ per a qualsevol $i < j$.

Definim inversi√≥, com el proc√©s d'intercanviar dos valors de la llista que no
estiguin en el seu ordre natural, es a dir, que el de l'esquerra sigui m√©s gran
que el de la dreta. Concretament, diem que $(A[i];A[j])$ es una inversi√≥, si
$A[i] > A[j]$ i $i < j$. 

Per exemple:
\
Si es t√© una llista **[1, 3, 2, 4]**, aleshores hi hauria una inversi√≥ **(3,2)**, ja que **3 > 2** i l'index de la posici√≥ del 3 √©s m√©s petit que la del 2. En altres paraules, es troba el 3 m√©s a l'esquerra de la llista que el 2. Per tant, nom√©s es produiria 1 inversi√≥.

### ‚úçÔ∏è Escriu una funci√≥ que calculi el nombre d'inversions dins una llista i que digui quines son.

```python
def comptador_inversions(llista):


    def merge_count(left, right):
        i = j = 0
        count = 0
        merged = []

        while i < len(left) and j < len(right):
            if left[i] <= right[j]:
                merged.append(left[i])
                i += 1
            else:
                merged.append(right[j])
                # Tots els left[i:] formen inversi√≥ amb right[j]
                count += len(left) - i
                j += 1

        merged.extend(left[i:])
        merged.extend(right[j:])
        return merged, count

    def sort_count(arr):
        if len(arr) < 2:
            return arr, 0
        
        mid = len(arr) // 2
        left, c1 = sort_count(arr[:mid])
        right, c2 = sort_count(arr[mid:])
        merged, c3 = merge_count(left, right)

        return merged, c1 + c2 + c3

    _, total_inversions = sort_count(llista)
    return total_inversions
```

```python
def comptador_inversions(llista):
    """
    Retorna nom√©s el nombre d'inversions.
    """

    def merge_count(left, right):
        i = j = 0                        # O(1)
        count = 0                        # O(1)
        merged = []                      # O(1)

        # Bucle principal del merge: com a molt recorre len(left)+len(right) = n  ‚Üí O(n)
        while i < len(left) and j < len(right):     # O(1) per iteraci√≥
            if left[i] <= right[j]:                 # O(1)
                merged.append(left[i])              # O(1)
                i += 1                              # O(1)
            else:
                merged.append(right[j])             # O(1)
                count += len(left) - i              # O(1)
                j += 1                              # O(1)

        # Afegeix el que resta (cost proporcional al nombre d‚Äôelements restants) ‚Üí O(n)
        merged.extend(left[i:])         # O(len(left)-i) ‚äÜ O(n)
        merged.extend(right[j:])        # O(len(right)-j) ‚äÜ O(n)

        return merged, count            # O(1)

    def sort_count(arr):
        if len(arr) < 2:                # O(1)
            return arr, 0               # O(1)

        mid = len(arr) // 2             # O(1)

        # Recursions: T(n/2) cadascuna
        left, c1 = sort_count(arr[:mid])     # O(n/2) per copiar + T(n/2)
        right, c2 = sort_count(arr[mid:])    # O(n/2) per copiar + T(n/2)

        # Merge cost: O(n)
        merged, c3 = merge_count(left, right) # O(n)

        return merged, c1 + c2 + c3      # O(1)

    # Crida inicial: T(n)
    _, total_inversions = sort_count(llista)   # O(n log n)

    return total_inversions               # O(1)


#COST TOTAL: O(n)
```
