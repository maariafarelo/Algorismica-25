## âœï¸ Problema 3.7: DescomposiciÃ³ en funcions

Escriu una funciÃ³ que permeti fer operacions de suma, resta i multiplicaciÃ³ amb nombres naturals 
a partir d'una entrada que sigui una cadena. L'usuari anirÃ  entrant les operacions, i quan vulgui 
acabar pitjarÃ  retorn i la cadena d'entrada quedarÃ  buida.

Exemple de funcionament:

&emsp;Entra operaciÃ³: 4+5\
&emsp;>9\
&emsp;Entra operaciÃ³: 7-2\
&emsp;>5\
&emsp;Entra operaciÃ³: 1*4\
&emsp;>4\
&emsp;Entra operaciÃ³:

```python

def calculadora():
    """
    Aquesta funciÃ³ representa una calculadora senzilla per a nombres naturals.
    Permet realitzar operacions de suma, resta i multiplicaciÃ³ a partir d'una
    entrada de cadena. L'execuciÃ³ acaba quan l'usuari prem retorn sense escriure res.

    Complexitat temporal total: O(1) per operaciÃ³ (O(k)), on k = longitud de la cadena.
    Complexitat espacial total: O(1)
    """

    while True:
        # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
        # 1. Llegir la cadena dâ€™entrada â†’ O(1)
        entrada = input("Entra operaciÃ³: ").strip()

        # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
        # 2. Si la cadena Ã©s buida, finalitzar â†’ O(1)
        if entrada == "":
            break

        # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
        # 3. Determinar lâ€™operador i separar operands â†’ O(k)
        #    (on k Ã©s la longitud de la cadena, mÃ xim 3-5 carÃ cters)
        if "+" in entrada:
            parts = entrada.split("+")   # O(k)
            operador = "+"
        elif "-" in entrada:
            parts = entrada.split("-")   # O(k)
            operador = "-"
        elif "*" in entrada:
            parts = entrada.split("*")   # O(k)
            operador = "*"
        else:
            print("> OperaciÃ³ no reconeguda")
            continue

        # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
        # 4. Convertir operands a enters â†’ O(1)
        try:
            a = int(parts[0])
            b = int(parts[1])
        except ValueError:
            print("> Error: entra nomÃ©s nombres naturals")
            continue

        # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
        # 5. Calcular resultat segons lâ€™operador â†’ O(1)
        if operador == "+":
            resultat = a + b
        elif operador == "-":
            resultat = a - b
        else:  # operador == "*"
            resultat = a * b

        # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
        # 6. Mostrar el resultat â†’ O(1)
        print(">", resultat)

```

## âœï¸ Problema 3.7: Nombre mÃ©s petit divisible

El nombre 2520 Ã©s el nombre mÃ©s petit que es pot dividir de manera exacta (sense decimals) per cada un dels nombres enters entre 1 i 10. Escriu un algorisme que calculi el nombre mÃ©s petit divisible per tots els nombres menors a un nombre $n$ donat.

```python
def mes_petit_divisible_per(n: int) -> int:
    """
    Aquesta funciÃ³ calcula el nombre mÃ©s petit divisible per tots els nombres menors o iguals a n.

    Exemple:
    >>> mes_petit_divisible_per(5)
    60  # (ja que Ã©s divisible per 1,2,3,4,5)

    Complexitat temporal total: O(n * log(min(a,b)))
    Complexitat espacial total: O(1)
    """

    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    # 1. ComprovaciÃ³ de validesa de n â†’ O(1)
    if n < 1:
        raise ValueError("n ha de ser un enter positiu")

    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    # 2. FunciÃ³ auxiliar per calcular el MCD amb lâ€™algorisme dâ€™Euclides
    def mcd(a: int, b: int) -> int:
        # Cada iteraciÃ³ redueix b segons el residu de la divisiÃ³,
        # per tant la complexitat Ã©s O(log(min(a,b)))
        while b != 0:                 # O(log(min(a,b)))
            a, b = b, a % b           # operacions O(1)
        return a

    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    # 3. Inicialitzar resultat â†’ O(1)
    resultat = 1

    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    # 4. Iterar de 2 fins a n i calcular el MCM acumulat â†’ O(n * log n)
    for i in range(2, n + 1):         # O(n) iteracions
        resultat = resultat * i // mcd(resultat, i)  # O(log(min(resultat, i)))

    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    # 5. Retornar resultat â†’ O(1)
    return resultat
```
