## âœï¸ Problema 3.16: Aritmetica modular. NIF

Escriu una funciÃ³ que, donat un nÃºmero en el format adient d'un NIF, validi si la lletra es correspon realment al nÃºmero del DNI. Has de fer servir alguna colÂ·lecciÃ³ de Python que sigui adequada. 

```python
def validar_NIF(cadenaNIF: str) -> bool:
    """
    Aquesta funciÃ³ valida si la lletra correspon al DNI.

    Complexitat temporal: O(1)
    Complexitat espacial: O(1)
    """

    # 1. ComprovaciÃ³ de longitud â†’ O(1)
    if len(cadenaNIF) != 9:
        return False

    # 2. Separar nÃºmero i lletra â†’ O(1)
    num_part = cadenaNIF[:-1]   # primers 8 carÃ cters
    lletra_dada = cadenaNIF[-1].upper()  # Ãºltim carÃ cter, en majÃºscula

    # 3. Validar que la part numÃ¨rica Ã©s realment un nÃºmero â†’ O(1)
    if not num_part.isdigit():
        return False

    # 4. Calcular la lletra que hauria de correspondre â†’ O(1)
    lletres = "TRWAGMYFPDXBNJZSQVHLCKE"  # colÂ·lecciÃ³ de lletres
    num_dni = int(num_part)
    lletra_correcta = lletres[num_dni % 23]

    # 5. Comparar la lletra donada amb la lletra correcta â†’ O(1)
    return lletra_dada == lletra_correcta
```

Repeteix el codi de la teva funciÃ³ i detalla els passos computacionals i l'ordre de complexitat amb O gran.

```python
def validar_NIF(cadenaNIF: str) -> bool:
    """
    Aquesta funciÃ³ valida si la lletra correspon al DNI.

    Complexitat temporal total: O(1)
    Complexitat espacial total: O(1)
    """

    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    # 1. ComprovaciÃ³ de longitud â†’ O(1)
    #    OperaciÃ³ constant, independent de la mida de l'entrada (sempre 9 carÃ cters).
    if len(cadenaNIF) != 9:
        return False

    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    # 2. Separar la part numÃ¨rica i la lletra â†’ O(1)
    #    L'operaciÃ³ d'slicing en un string petit i fix (9 carÃ cters) Ã©s constant.
    num_part = cadenaNIF[:-1]        # O(1)
    lletra_dada = cadenaNIF[-1].upper()  # O(1)

    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    # 3. Validar que la part numÃ¨rica Ã©s realment un nÃºmero â†’ O(1)
    #    `.isdigit()` recorre la cadena, perÃ² com que tÃ© longitud constant (8 carÃ cters),
    #    considerem el cost constant.
    if not num_part.isdigit():
        return False

    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    # 4. Calcular la lletra corresponent â†’ O(1)
    #    - CreaciÃ³ o accÃ©s a la cadena de lletres: O(1)
    #    - CÃ lcul del mÃ²dul (%): O(1)
    #    - IndexaciÃ³ per posiciÃ³: O(1)
    lletres = "TRWAGMYFPDXBNJZSQVHLCKE"  # O(1)
    num_dni = int(num_part)              # conversiÃ³ a enter â†’ O(1)
    lletra_correcta = lletres[num_dni % 23]  # cÃ lcul i accÃ©s â†’ O(1)

    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    # 5. Comparar la lletra calculada amb la lletra donada â†’ O(1)
    #    ComparaciÃ³ simple entre dos carÃ cters.
    return lletra_dada == lletra_correcta  # O(1)
```

## âœï¸ Problema 3.16: Aritmetica modular. Fulla de cÃ lcul

Escriu una funciÃ³ per convertir un nombre en el nom de la columna del full de cÃ lcul corresponent.

Ex. 1 => A, 11 => K

```python
def conversio_fulla_calcul(num: int) -> str:
    """
    Aquesta funciÃ³ tradueix el valor num en el nom corresponent que tindrÃ­em
    en un full de cÃ lcul (A, B, ..., Z, AA, AB, ...).

    Complexitat temporal total: O(logâ‚â‚‚â‚†â‚(num))
    Complexitat espacial total: O(logâ‚â‚‚â‚†â‚(num))
    """

    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    # 1. ComprovaciÃ³ de validesa del nÃºmero â†’ O(1)
    if num <= 0:
        raise ValueError("El nÃºmero ha de ser positiu")

    resultat = ""  # cadena buida per construir el nom de la columna

    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    # 2. Bucle de conversiÃ³ (aritmÃ¨tica modular) â†’ O(logâ‚â‚‚â‚†â‚(num))
    #    En cada iteraciÃ³ es divideix el nÃºmero per 26,
    #    aixÃ­ que el nombre de passos Ã©s proporcional al log base 26 del nÃºmero.
    while num > 0:
        num -= 1                          # ajust per passar de base 1 a base 0 â†’ O(1)
        residu = num % 26                 # operaciÃ³ mÃ²dul â†’ O(1)
        lletra = chr(65 + residu)         # converteix 0â†’A, 1â†’B, ... 25â†’Z â†’ O(1)
        resultat = lletra + resultat      # concatenaciÃ³ â†’ O(1) (cadena petita)
        num //= 26                        # divisiÃ³ entera â†’ O(1)

    # â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    # 3. Retorn del resultat â†’ O(1)
    return resultat
```
