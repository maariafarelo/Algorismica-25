Escriu una funci贸 que faci la divisi贸 entera entre dos nombres enters, positius i m茅s grans que $1$. No utilitzis els operadors *, /, % per resoldre el problema. 

```python
def divisio_entera(dividend: int, divisor: int) -> int:
    """
    Calcula la divisi贸 entera de dos nombres naturals sense utilitzar
    els operadors *, /, %.

    Parameters
    ----------
    dividend : int
    divisor : int

    Returns
    -------
    quocient : int
    """
    if divisor == 0:
        raise ValueError("Divisi贸 per zero no permesa")

    quocient = 0
    resta = dividend

    while resta >= divisor:
        resta -= divisor
        quocient += 1

    return quocient
```
