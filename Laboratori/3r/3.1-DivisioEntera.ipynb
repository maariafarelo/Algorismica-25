## âœï¸ Problema 3.1: DivisiÃ³ entera - Maria Farelo Iglesias

Escriu una funciÃ³ que faci la divisiÃ³ entera entre dos nombres enters, positius i mÃ©s grans que $1$. No utilitzis els operadors *, /, % per resoldre el problema. 

```python
def divisio_entera(dividend: int, divisor: int) -> int:
    """
    Calcula la divisiÃ³ entera de dos nombres naturals sense utilitzar
    els operadors *, /, %.

    Parameters
    ----------
    dividend : int
    divisor : int

    Returns
    -------
    quocient : int
    """
    if divisor == 0:
        raise ValueError("DivisiÃ³ per zero no permesa")

    quocient = 0
    resta = dividend

    while resta >= divisor:
        resta -= divisor
        quocient += 1

    return quocient
```

```python
# ðŸ”¬  per provar...

assert divisio_entera(10,2) == 5
assert divisio_entera(20,10) == 2
assert divisio_entera(9,2) == 4
```

### ðŸ“Š CÃ lcul complexitat

Repeteix el codi de la teva funciÃ³ i detalla els passos computacionals i l'ordre de complexitat amb O gran.

```python
def divisio_entera(dividend: int, divisor: int) -> int:
    """
    Calcula la divisiÃ³ entera de dos nombres naturals sense utilitzar
    els operadors *, /, %.

    Complexitat temporal: O(N / D)
    Complexitat espacial: O(1)
    On N = dividend, D = divisor
    """

    # 1. ComprovaciÃ³ de divisiÃ³ per zero â†’ O(1)
    if divisor == 0:
        raise ValueError("DivisiÃ³ per zero no permesa")

    # 2. InicialitzaciÃ³ de variables â†’ O(1)
    quocient = 0
    resta = dividend

    # 3. Bucle principal:
    #    S'executa tantes vegades com cops es pot restar 'divisor' de 'dividend'
    #    Nombre d'iteracions â‰ˆ N / D
    while resta >= divisor:               # ComparaciÃ³ â†’ O(1) per iteraciÃ³
        resta -= divisor                  # Resta â†’ O(1)
        quocient += 1                     # Increment â†’ O(1)
        # Total per iteraciÃ³: O(1)

    # 4. Retorn del resultat â†’ O(1)
    return quocient
```
