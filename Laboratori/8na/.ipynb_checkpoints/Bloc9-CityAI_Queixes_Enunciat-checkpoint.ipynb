## ‚úçÔ∏è Exercici 1: Adaptaci√≥ de mergesort

Completa el codi del mergesort amb una variaci√≥: usarem una funci√≥ lambda "key" com a criteri d'ordenaci√≥.

A continuaci√≥ us copiem el codi original i el codi adaptat on falten algunes parts.

```python
# Codi original de la funci√≥ merge

def merge(x:list[any],y:list[any])->list[any]:
    """
    Fusiona dues llistes pr√®viament ordenades en una nova llista ordenada
    :param
    x: una llista ordenada
    y: una llista ordenada
    :return
    una nova llista amb tots els elements d'x i y, ordenats
    """
    # print("MERGE: m'han cridat amb",x,y)  # xivato
    if len(x) < 1:
        return y
    if len(y) < 1:
        return x
    llista_nova = []
    i = j = 0
    while i < len(x) and j < len(y):
        if x[i] <= y[j]:
            llista_nova.append(x[i])
            i += 1
        else:
            llista_nova.append(y[j])
            j += 1
    llista_nova.extend(x[i:])
    llista_nova.extend(y[j:])
    return llista_nova
```

```python
# Codi original de la funci√≥ mergesort

def mergesort(l:list[any])->list[any]:
    """
    Ordena una llista
    :param
    l1: una llista a ordenar
    :return
    una nova llista ordenada
    """
    # print("MERGESORT: m'han cridat amb",l)  # xivato
    if len(l) < 2:
        return l
    else:
        meitat = len(l) // 2
        esquerra = mergesort(l[:meitat])
        dreta = mergesort(l[meitat:])
        return merge(esquerra, dreta)
```
