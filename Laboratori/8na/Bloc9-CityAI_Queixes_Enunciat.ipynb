## âœï¸ Exercici 1: AdaptaciÃ³ de mergesort

Completa el codi del mergesort amb una variaciÃ³: usarem una funciÃ³ lambda "key" com a criteri d'ordenaciÃ³.

```python
# Codi original de la funciÃ³ merge

def merge(x:list[any],y:list[any])->list[any]:
    """
    Fusiona dues llistes prÃ¨viament ordenades en una nova llista ordenada
    :param
    x: una llista ordenada
    y: una llista ordenada
    :return
    una nova llista amb tots els elements d'x i y, ordenats
    """
    if len(x) < 1:
        return y
    if len(y) < 1:
        return x
    llista_nova = []
    i = j = 0
    while i < len(x) and j < len(y):
        if x[i] <= y[j]:
            llista_nova.append(x[i])
            i += 1
        else:
            llista_nova.append(y[j])
            j += 1
    llista_nova.extend(x[i:])
    llista_nova.extend(y[j:])
    return llista_nova
```

## âœï¸ Exercici 3: Mostrar resultats

Escriu una funciÃ³ que donada una llista de queixis la mostri per pantalla de manera ordenada i clara.
La funciÃ³ ha de permetre indicar quants Ã­tems mostrar del principi i quants del final.

Utilitza la plantilla donada.

```python
def mostra_queixes(queixes: list[dict], n_inici: int, n_final: int) -> None:
    def format_data(q):
        dia = q.get("DIA_DATA_ALTA")
        mes = q.get("MES_DATA_ALTA")
        any = q.get("ANY_DATA_ALTA")
        if dia and mes and any:
            return f"{dia:02d}/{mes:02d}/{any}"
        else:
            return "Sense data"

    def format_districte(q):
        d = q.get("DISTRICTE")
        return d if d and str(d).lower() != "nan" else "Sense districte"

    print("ðŸ“Œ Primeres queixes:")
    for q in queixes[:n_inici]:
        print(f"- ID: {q.get('FITXA_ID')} | Districte: {format_districte(q)} | "
              f"Tipus: {q.get('TIPUS')} | Data: {format_data(q)}")

    print("\nðŸ“Œ Ãšltimes queixes:")
    for q in queixes[-n_final:]:
        print(f"- ID: {q.get('FITXA_ID')} | Districte: {format_districte(q)} | "
              f"Tipus: {q.get('TIPUS')} | Data: {format_data(q)}")
```
