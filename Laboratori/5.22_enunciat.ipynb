### 5.22 Ordenar llista aparellada

Suposem que necessitem ordenar llistes els nombres de les quals apareixen duplicats, un darrere l'altre. Per exemple, la llista [1, 1, 4, 4, 3, 3, 5, 5]. Prova l' algorisme d'ordenació mergesort. Calcula la complexitat d'aquest algorisme a nivell de comparacions i d'intercanvi d'elements.

No pots utilitzar funcions d'ordenació de Python.

El mergesort ha de retornar la llista modificada.

```python
def mergesort_parelles(llista):

    def merge(left, right):
        i = j = 0
        result = []

        while i < len(left) and j < len(right):
            # COMPARACIÓ → 1 comparació
            if left[i] <= right[j]:
                result.append(left[i])  # MOVIMENT (còpia)
                i += 1
            else:
                result.append(right[j]) # MOVIMENT (còpia)
                j += 1

        # MOVIMENTS (còpies) restants
        result.extend(left[i:])
        result.extend(right[j:])
        return result

    # Cas base
    if len(llista) <= 1:
        return llista

    # Divisió
    mid = len(llista) // 2
    left = mergesort_parelles(llista[:mid])
    right = mergesort_parelles(llista[mid:])

    # Fusió
    return merge(left, right)
```

```python
llista = [1, 1, 7, 7, 2, 2, 3, 3, 5, 5]
assert mergesort_parelles(llista) == [1, 1, 2, 2, 3, 3, 5, 5, 7, 7]
```
