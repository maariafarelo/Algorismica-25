### ‚úçÔ∏è Problema 5.1 Suma Llista

Escriu una **funci√≥ recursiva** `suma_llista(llista)` que calculi la **suma de tots els elements** d‚Äôuna llista d‚Äôenters.

---

#### üß† Tasques a realitzar

1. Implementa la funci√≥ `suma_llista(llista)` fent √∫s exclusiu de **recursivitat** (sense bucles `for` o `while`).  
2. Identifica clarament quin √©s el **cas base** de l‚Äôalgorisme.

```python
def suma_llista(llista:list[int]):
    """
    Calcula de forma recursiva la suma de tots els elements d'una llista num√®rica.

    Parameters
    ----------
    llista : list
        Llista d‚Äôelements num√®rics (int o float).

    Returns
    -------
    int | float
        La suma total de tots els elements de la llista.
    """
    #Cas BASE:
    if not llista:
        return 0

    #Altres casos:
    return llista[0] + suma_llista(llista[1:])
```

```python
assert suma_llista([1,2,3,4]) == 10
assert suma_llista([10,20,30,40]) == 100
assert suma_llista([-1,-2,0,1,2]) == 0
assert suma_llista ([-1,-5,5,-6,2]) == -5
```

### üìä C√†lcul complexitat

Repeteix el codi de la teva funci√≥ i detalla els passos computacionals i l'ordre de complexitat amb O gran. 

```python
def suma_llista(llista: list[int]) -> int:
    # Cas base: si la llista est√† buida, retorna 0
    if not llista:  # O(1)
        return 0  # O(1) 

    # Cas recursiu: retorna l'element actual (llista[0]) + la suma de la resta de llista (llista[1:])
    return llista[0] + suma_llista(llista[1:])  # O(n)


#ORDRE COMPLEXITAT: Al ser recursiu hem de recorre tota la llista, per tant √©s O(n)
```
