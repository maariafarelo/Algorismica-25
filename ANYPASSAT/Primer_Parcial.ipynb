     nombre_bits = 0

```python
def primers_Wieferich(n = 5000):
    """
    Aquesta funció calcula els primer de Wieferich menors que n = 5000.

    Parameters
    ----------
    n: int

    Returns
    -------
    wieferich: list
    """
    l = []

    for i in range (2,n):
        a=2**(i-1)-1
        b=(i**2)
        c= a % b
        if ( c == 0 ):
            l.append(i)

    return l

```
