### Exercici 1 (2 punts)

Un primer de Wieferich és un nombre primer p tal que $p^2$ divideix $2^{(p-1)}-1$. Quins primers de Wieferich hi ha menors que 5000?

Per exemple: 1093, ja que com es veu al codi de sota compleix la condició.


```python
a=2**(1093-1)-1
print(a)
b=(1093**2)
print(b)
c= a % b
print(c)
```

Calcula els passos computacionals de primers_Weiferich i l'ordre O gran de complexitat. Explica què es la n.

```python

def primers_Wieferich(n=5000):
    wieferich = []                     # Paso 1:  O(1) estas creando una lista vacia
    for p in range(2, n+1):            # Paso 2: O(n) iteracions
        if pow(2, p-1, p**2) == 1:    # Paso 3: O (log p) multiplicació modular i comparació O(1)
            wieferich.append(p)        # Paso 4: O(1) afegir un element a la llista
    return wieferich                    # Paso 5: O(1)

# Complexitat = 0(n) ja que mirem tots els números fins O(n+1) és a dir O(n) de complexitat espacial
```

```python
assert primers_Wieferich(5000) == [1093, 3511]
```

Per facilitar-vos les coses hem convertit aquesta informació en diccionaris:

1. Crea una funció "recomptelletres" que donat un fitxer de text computi quantes vegades apareix cadascun dels caràcters del text i les retorni en un diccionari. També ha de retornar el total de lletres del text

2. Crea una segona funció "frequencialletres" que, amb una crida a la funció recomptelletres, compti la freqüència de les lletres a a j.

3. Crea una tercera funció "calculardiferencia" que donats dos diccionaris amb les freqüències de les lletres a a j,  compari la diferència de la frequència de cada lletra, $i$, en el text, $F_i^{t}$,  i la freqüència en la llengua, $F_i^{l}$, segons la fòrmula següent.

$E_l = \sum_{i="a"}^{"j"} {(F_i^{t} - F_i^{l})^2}$


4. Crea una darrera funció "idioma" que donat un fitxer, identifiqui en quin idioma ha estat escrit.




```python
langles = { "a":8.167, "b":1.492, "c":2.782, "d": 4.253, "e": 12.702, "f": 2.228, "g":2.015, "h": 6.094, "i": 6.966, "j": 0.153}

lfrances = { "a":7.636, "b":0.901, "c":3.26, "d": 3.669, "e":14.715, "f":1.066,  "g":0.866, "h":0.737, "i":7.529, "j":0.613}

lespanyol = { "a":11.525, "b":2.215, "c":4.019, "d": 5.51, "e":12.681, "f":0.692, "g":1.768, "h":0.703, "i":6.247, "j":0.443}
```

```python
langles = { "a":8.167, "b":1.492, "c":2.782, "d": 4.253, "e": 12.702, "f": 2.228, "g":2.015, "h": 6.094, "i": 6.966, "j": 0.153}

lfrances = { "a":7.636, "b":0.901, "c":3.26, "d": 3.669, "e":14.715, "f":1.066,  "g":0.866, "h":0.737, "i":7.529, "j":0.613}

lespanyol = { "a":11.525, "b":2.215, "c":4.019, "d": 5.51, "e":12.681, "f":0.692, "g":1.768, "h":0.703, "i":6.247, "j":0.443}


def recomptelletres(fitxer):
    """
    Aquesta funció retorna el recompte de les lletres en un text, i el total de lletres.
    
    Parameters
    ----------
    fitxer: string
        text a codificar
        
    Returns
    -------
   rec: diccionari
        diccionari amb el recompte de les lletres
    total: int
        total de lletres  
    """
    recompte = {c: 0 for c in "abcdefghij"}
    total = 0
    with open(fitxer,'r', encoding="utf-8") as fitxero:
        for linea in fitxero:
            for car in linea.lower():
                if car in recompte:
                    recompte[car] += 1
                    total +=1
    return recompte, total

def frequencialletres(fitxer):
    """
    Aquesta funció retorna la freqüència de les lletres a a j en un text.
    
    Parameters
    ----------
    fitxer: string
        text a codificar
        
    Returns
    -------
    frec: diccionari
        diccionari amb la freqüència de les lletres a a j
    """
    recompte, total = recomptelletres(fitxer)
    frec = {}                     # Creamos un diccionario vacío
    for c in recompte:            # Recorremos cada letra
        if total > 0:             # Comprobamos que no sea 0
            frec[c] = recompte[c] / total * 100
        else:
            frec[c] = 0

    return frec


def calculardiferencia(freq1, freq2):
    """
    Aquesta funció retorna la diferència de freqüències entre els dos diccionaris entrats.
    freq1 i freq2 són diccionaris amb la freqüència de les lletres a a j.
    
    Parameters
    ----------
    freq1, freq2: diccionari
        diccionaris amb la freqüència de les lletres a a j
        
    Returns
    -------
    diferencia: float
        diferència entre les freqüències
        segons la fòrumula de l'enunciat
    """
    diferencia = 0
    for c in freq1:
        diferencia += (freq1[c] - freq2[c])**2
    return diferencia
    

def idioma(fitxero):
    """
    Aquesta funció retorna l'idioma estimat d'un text.
    
    Parameters
    ----------
    fitxer: string
        text a codificar
        
    Returns
    -------
    idioma: string
        idioma estimat del text
    """
    freq_text = frequencialletres(fitxero)
    dif_ingles = calculardiferencia(freq_text, langles)
    dif_frances = calculardiferencia(freq_text, lfrances)
    dif_espanyol = calculardiferencia(freq_text, lespanyol)
    
    difs = {"anglès": dif_ingles, "francès": dif_frances, "espanyol": dif_espanyol}
    return min(difs, key=difs.get)
```

```python
assert idioma("hamlet.txt") == "anglès"
assert idioma("quijote.txt") == "espanyol"
assert idioma("miserables.txt") == "francès"
```
