#### T1.1: Quants bits té com a màxim la suma de x+y, si ambdós tenen n bits?

Quan sumem dos nombres binaris, cadascun amb **n bits**, la suma pot tenir fins a **n+1 bits** com a màxim. Això és perquè, si ambdós nombres són els més grans possibles (tots els bits iguals a 1), es pot generar un bit de desbordament (carry) que augmenta la longitud de la suma en un bit addicional.

Exemple: Sumar 111 (7 en decimal) + 111 (7 en decimal) resulta en 1110 (14 en decimal), que té un bit més que els operands.

Per tant, la suma de dos nombres de **n bits** pot tenir com a màxim **n+1 bits**.



#### T1.2: Podem considerar de la mateixa complexitat una operació amb nombres decimals que amb nombres binaris?

No, les operacions amb nombres decimals no tenen la mateixa complexitat que les operacions amb nombres binaris.

Els ordinadors treballen nativament amb binari (0s i 1s), i les operacions amb binari són més simples a nivell de maquinari. En canvi, el sistema decimal (base 10) requereix conversions addicionals o aritmètica més complexa per ser processat per l'ordinador, la qual cosa introdueix una sobrecàrrega en temps de computació.

En resum, les operacions **binàries són més eficients** per als ordinadors.



#### T1.3: En aritmètica modular com podem simplificar x^25?

En aritmètica modular, podem utilitzar el concepte d'exponenciació eficient per reduir el càlcul d'exponencials elevades com **x^25 mod n**. La idea és descompondre l'exponent de manera que es pugui calcular en menys passos utilitzant l'exponenciació per quadrats.

Per exemple, podem expressar **x^25** com:

$$ x^{25} = x^{16} \cdot x^8 \cdot x $$

I, mitjançant **exponenciació per quadrats**, podem calcular cadascuna d'aquestes potències modulars per separat, simplificant així el càlcul. Aquesta tècnica redueix significativament la quantitat d'operacions necessàries per calcular exponencials elevades.



#### T1.4: Quina complexitat té l'exponenciació modular?

L'exponenciació modular es pot realitzar de manera eficient utilitzant l'algorisme d'**exponenciació per quadrats** (també conegut com a "square and multiply"). Aquest algorisme redueix la complexitat a **O(log e)**, on **e** és l'exponent.

Això és molt més eficient que una aproximació directa, que tindria una complexitat **O(e)**, ja que es reduiria a fer un nombre lineal de multiplicacions. L'exponenciació per quadrats aprofita la descomposició de l'exponent en potències de 2, fent el procés molt més ràpid.

Així doncs, la **complexitat de l'exponenciació modular és O(log e)**.

#### T2.2: Quins són els errors més difícils de detectar automàticament (per un debugger o similar)?

Els errors més difícils de detectar automàticament són:

- **Errors d'ús de memòria (p.ex., fugues de memòria):** Tot i que algunes eines poden ajudar, detectar l'ús ineficient o incorrecte de memòria és complex en grans aplicacions, especialment en llenguatges de baix nivell com C/C++.
- **Errors d'execució en temps específics (errors temporals):** Errors que només es manifesten en condicions específiques d'execució o a causa de l'ordre dels esdeveniments.
- **Bugs d'estat global:** Quan l'estat global d'un programa es modifica de manera incorrecta, pot ser difícil seguir el rastre d'aquest tipus d'errors, especialment en aplicacions grans.

#### T2.3: Perquè a Python mai no anomenarem a una variable "list"?

A Python, **`list`** és una paraula reservada que representa el **tipus de dades incorporat (builtin)** per a les llistes. Si anomenem una variable **`list`**, estarem sobreescrivint el nom del tipus incorporat, fet que impedirà utilitzar la funcionalitat nativa de llistes. Això pot provocar errors inesperats en el codi quan intentem cridar o crear una llista utilitzant aquest nom sobreescrit.

Per tant, és una mala pràctica i és recomanable evitar anomenar variables amb noms de tipus incorporats com `list`, `dict`, `int`, etc.

#### Ineficiencia del Cálculo Recursivo del Número de Fibonacci

El siguiente código implementa una función para calcular el **n-ésimo número de Fibonacci** de manera recursiva. Aunque es funcional, esta implementación es **altamente ineficiente** por varias razones, que se detallan a continuación.

#### 1. Duplicación de Cálculos (Redundancia Recursiva)

La función `fib1` calcula el valor de Fibonacci para cada número en una secuencia descendente hasta llegar a los casos base (0 y 1). En una llamada a `fib1(n)`, se hacen llamadas recursivas a `fib1(n-1)` y `fib1(n-2)`. Sin embargo, cada llamada recursiva, a su vez, recalcula los mismos subproblemas en niveles inferiores de la recursión.

Esto genera un **crecimiento exponencial** en el número de llamadas recursivas, ya que el mismo valor de Fibonacci se calcula múltiples veces.

##### Ejemplo de Redundancia

Supongamos que `fib1(5)` hace una llamada a `fib1(4)` y `fib1(3)`. A continuación, `fib1(4)` hace llamadas a `fib1(3)` y `fib1(2)`. Observe que `fib1(3)` se llama tanto por `fib1(5)` como por `fib1(4)`, recalculando así los mismos valores. Esta duplicación de cálculos se repite para cada número de Fibonacci en la secuencia, haciendo que el algoritmo sea exponencialmente lento.

#### 2. Complejidad Temporal Exponencial

La complejidad temporal de `fib1` es **exponencial**: \(O(2^n)\).

Esto significa que el tiempo de ejecución crece exponencialmente a medida que aumenta \(n\), lo que hace que el algoritmo se vuelva intractable muy rápidamente para valores grandes de \(n\). Por ejemplo, para \(n = 40\), se realizan más de un millón de llamadas recursivas, lo que hace que la función tarde mucho tiempo en resolver.

#### 3. Alternativa Más Eficiente: Memorización (Memoization) o Cálculo Iterativo

Una manera de mejorar la eficiencia es utilizar **memoization** para almacenar los resultados de subproblemas ya calculados y reutilizarlos en lugar de recalcularlos. Otra opción es utilizar una solución **iterativa** o **programación dinámica**, que calcula los números de Fibonacci de manera secuencial, evitando la redundancia.

#### Ejemplo de Implementación Eficiente Iterativa

def fib_iter(n):
    if n <= 1:
        return n
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b

Diem que dues cadenes de caràcters a i b són isomorfes si les ocurrències de cada caràcter de la cadena a es poden reemplaçar per un caràcter de manera que la convertim en la cadena b, i viceversa.

Per exemple, les dues cadenes:

+ a = "Teulades"
+ b = "Piscines"

són isomorfes perquè podem establir aquestes correspondències:

T <=> P, E <=> I, U <=> S, L <=> C, D <=> N

cal tenir en compte que un caràcter pot tenir una correspondència amb si mateix, però dos caràcters diferents no poden correspondre a un mateix caràcter.

Escriu una funció anomenada "isomorf" tal que, donades dues cadenes de caràcters d'entrada, retorni "True" si són isomorfes i "False" en cas contrari.

Fes servir la plantilla següent:

````python
"""
Plantilla:
s = "CBAABC"
t = "DEFFED"
l = []
l2 = []
dic2 = {}
for i in s:
  if i not in l:
    l.append(i)
for x in t:
  if x not in l2:
    l2.append(x)
if(len(l) != len(l2)):
  print(False)
else:
  for m in range(len(l)):
    dic2[l[m]] = l2[m]
print(dic2)
"""
````

````python
def isomorf(cadX, cadY):
    dic2 = {}  # Diccionario para guardar los mapeos
    # Si las longitudes de las cadenas no son iguales, no pueden ser isomorfas
    if len(cadX) != len(cadY):
        return False
    else:
        # Recorremos ambas cadenas y comprobamos los mapeos
        for i in range(len(cadX)):
            if cadX[i] in dic2:
                # Si el carácter de cadX ya tiene un mapeo, verificamos si coincide
                if dic2[cadX[i]] != cadY[i]:
                    return False  # Si no coincide, no son isomorfos
            else:
                # Si el carácter de cadX no tiene un mapeo, lo asignamos
                dic2[cadX[i]] = cadY[i]
        
        # Si todos los mapeos fueron consistentes, las cadenas son isomorfas
        return True
````

````python
assert isomorf('CBAABC', 'DEFFED') == True
assert isomorf('XXX', 'YYY') == True
assert isomorf('RAMBUNCTIOUSLY', 'THERMODYNAMICS') == True
assert isomorf('XXY', 'XYY') == False
assert isomorf('ABAB', 'CD') == False
````

+ 3. n = k! + 1 per a algun k

````python
import math

def numero_3_condicions():
    """
    Aquesta funció troba els números que compleixen les tres condicions demanades
    
    Returns
    -------
    buscat: int
    """
    l = []
    l2 = []
    f = set()
    
    # Generem k! + 1 i afegim a l'set
    for i in range(20):  # Ampliem el rang per assegurar que capturem suficients valors
        f.add(math.factorial(i) + 1)
    
    # Convertim l'set a llista i filtrem els que són menors de 10,000
    for i in f:
        if i < 10000:
            l.append(i)
    
    # Comprovem si són quadrats perfectes
    for n in l[:]:  # Fem còpia de la llista per evitar modificar mentre recorrem
        sqrt_n = int(math.sqrt(n))
        if sqrt_n * sqrt_n == n:  # Si és un quadrat perfecte
            l2.append(n)
    print(l2)

    # Si no trobem cap número vàlid, retornem None
    if not l2:
        return None

    # Busquem el màxim de l2
    buscat = max(l2)

    return buscat

# Crida de la funció i imprimir el resultat
resultat = numero_3_condicions()
````

````python
def processar_matriu(llista_de_llistes):
    """
    Aquest programa retorna una matriu de la mateixa mida que la que rep per paràmetre, 
    amb els valors transformats: cada element passa a tenir com a valor el promig del seu 
    valor original i del valors dels seus veins.
    Es consideren veins aquells elements que comparteixen una aresta amb la casella.
       
    Parameters
    ----------
    llista_de_llistes: una matriu no buida d'enters o reals expressada com una llista de llistes
    
    Returns
    -------
    matriu_transformada: una matriu no buida d'enters o reals expressada com una llista de llistes
    """
    # Obtenim les dimensions de la matriu
    files = len(llista_de_llistes)
    columnes = len(llista_de_llistes[0]) if files > 0 else 0

    # Inicialitzem la matriu result amb zeros
    matriu_transformada = [[0] * columnes for _ in range(files)]
    
    # Definim els desplaçaments per accedir als veïns (amunt, avall, esquerra, dreta)
    desplaçaments = [(-1, 0), (1, 0), (0, -1), (0, 1)]  

    # Iterem sobre cada casella de la matriu
    for i in range(files):
        for j in range(columnes):
            suma = llista_de_llistes[i][j]  # Comencem amb el valor original
            comptador = 1  # Comptador inicialitzat amb 1 per a l'original
            
            # Iterem sobre cada veí possible
            for dx, dy in desplaçaments:
                ni, nj = i + dx, j + dy  # Coordenades del veí
                if 0 <= ni < files and 0 <= nj < columnes:  # Comprovem que el veí està dins de la matriu
                    suma += llista_de_llistes[ni][nj]  # Afegim el valor del veí
                    comptador += 1  # Augmentem el comptador
            
            # Calculem el promig
            matriu_transformada[i][j] = suma / comptador
    
    return matriu_transformada
````
