```python
def conversio_color_base(n):
    """
    Converteix un número en la seva representació en colors en base 4.
    
    Parameters
    ----------
    n: int
    
    Returns
    -------
    result: str
        Representació del número en colors
    """
    color_map = {0: "Vermell", 1: "Blau", 2: "Verd", 3: "Groc"}
    result = ""
    
    # Sempre afegim almenys un color
    primer_color = color_map[n % 4]
    result = primer_color
    
    n = n // 4
    while n > 0:
        resta = n % 4
        result = color_map[resta] + result
        n = n // 4
    
    return result

```

```python
def suma_digits(num, potencia):
    """
    Calcula la suma dels dígits de num elevat a potencia.
    
    Parameters
    ----------
    num: int
    potencia: int
    
    Returns
    -------
    suma: int
    """
    numero = num ** potencia
    suma = 0

    while numero > 0:
        suma += numero % 10
        numero = numero // 10

    return suma
```
