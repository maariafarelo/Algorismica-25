```python
def conversio_color_base(n):
    color_map = {0: "Vermell", 1: "Blau", 2: "Verd", 3: "Groc" }
    
    result= []                                   #1
    result = color_map[n%4]                      #1+1+1
    while n>=4:                                  #log(num, 4)
        n//=4                                    # 1+1
        resta = n % 4                            # 1+1
        result= color_map[resta] + result        # 1+1+1
    return result                                #1


# Complexitat O(log)
```
