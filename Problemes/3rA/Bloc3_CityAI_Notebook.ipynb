```python
import csv
from collections import defaultdict

def mitjana_districte() -> list[int]:
    '''  
    Retorna una llista amb la mitjana de renda per barri de Barcelona,
    mostrant pas a pas el que est√† passant.
    '''

    # Diccionaris per acumular suma de rendes i nombre de registres per barri
    suma_rendes = defaultdict(int)
    comptadors = defaultdict(int)

    # Obre el CSV
    with open("04-2022_sense_comes_renda_disponible_llars_per_persona.csv", encoding="utf-8") as fitxer:
        lector = csv.reader(fitxer, delimiter=';')

        for i, fila in enumerate(lector):
            barri = fila[4]        # camp del barri
            renda = int(fila[6])   # camp de la renda

            # Suma i comptador per al barri
            suma_rendes[barri] += renda
            comptadors[barri] += 1

    # Calculem la mitjana per barri
    mitjanes = []
    for barri in suma_rendes:
        mitjana = suma_rendes[barri] / comptadors[barri]
        print(f"Mitjana final {barri}: {mitjana:.2f}")
        mitjanes.append(int(mitjana))   # convertim a int per retornar list[int]

    return mitjanes


# Cridem la funci√≥
llista_mitjanes = mitjana_districte()

# Mostrem la llista final de mitjanes
print("Llista de mitjanes per barri:", llista_mitjanes)


```

```python
# Funci√≥ per trobar el barri amb menor renda
def barri_menor_renda() -> str:
    '''Retorna el barri amb la renda m√©s baixa.'''

    # Diccionaris per acumular suma de rendes i comptadors per barri
    suma_rendes = defaultdict(int)
    comptadors = defaultdict(int)

    # Obrim el CSV amb les dades de renda
    with open("04-2022_sense_comes_renda_disponible_llars_per_persona.csv", encoding="utf-8") as fitxer:
        lector = csv.reader(fitxer, delimiter=';')
        for fila in lector:
            barri = fila[4]
            renda = int(fila[6])
            suma_rendes[barri] += renda
            comptadors[barri] += 1

    # Calculem la mitjana de renda per barri
    mitjanes = {}
    for barri in suma_rendes:
        mitjanes[barri] = suma_rendes[barri] / comptadors[barri]

    # Retornem el barri amb la renda m√≠nima
    barri_min = min(mitjanes, key=mitjanes.get)
    return barri_min
```

```python
# Funci√≥ d'exponenciaci√≥ per quadrats
def calcul_piramidal(base:int, exponent:int) -> int:
    '''
    Calcula base^exponent eficientment (exponenciaci√≥ per quadrats).
    
    :param base: nombre base
    :param exponent: exponent >= 0
    :return: base^exponent
    '''
    resultat = 1
    while exponent > 0:
        if exponent % 2 == 1:       # si exponent √©s imparell
            resultat *= base
        base *= base
        exponent //= 2
    return resultat
```
