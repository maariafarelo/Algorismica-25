En aquest notebook es reproduiran les tres versions vistes a teoria de Fibonacci i per cadascuna es mostraran les crides que es fan, i es calcular√† el cost i l'ordre de complexitat O gran.

## üß† Versi√≥ recursiva

````python
def fibonacci(n: int) -> int:
    """
    Calcula el n-√®ssim nombre de Fibonacci de forma recursiva.

    :param n: √çndex (n ‚â• 0) del nombre de Fibonacci a calcular.
    :return: El nombre de Fibonacci a la posici√≥ n.
    """
    print("Estem cridant la funci√≥ amb el valor ",n)
    if n==0:
        return n
    if n==1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)
````

````python
fibonacci(5)
````

### ‚è±Ô∏è C√†lcul del cost

Si n <= 1 fem una crida, el cost √©s constant O(1)

si n>1 la funci√≥ fa dues crides m√©s un cost constant (comparacions i suma)
$T(n)=T(n‚àí1)+T(n‚àí2)+C$ i aix√≤ √©s repetir√† fins arribar al cas base. Aix√≤ ho podem representar amb un arbre **binari**, en el que cada fulla de l‚Äôarbre de crides √©s un **cas base** (`n = 0` o `n = 1`).

I quantes fulles hi ha? el nombre de fulles √©s aproximadament **F(n)** (nombre de Fibonacci), que creix com:
  $$
  F(n) \approx \frac{\varphi^n}{\sqrt{5}}
  $$
  on $\varphi \approx 1.618$.

El cost de l'algorisme √©s proporcional al nombre total de nodes de l‚Äôarbre.
- En un arbre binari ple:
  $$
  \text{Nodes totals} \approx 2 \times \text{Fulles}
  $$
- Si treiem les constants multiplicatives, en notaci√≥ Big-O, ens queda un cost de:
  $$
  T(n) = O(\varphi^n) \quad \text{aproximadament } O(2^n)
  $$

I a banda hem gastat molta mem√≤ria perqu√® la pila d'execuci√≥ ha tingut tamb√© totes aquestes crides.


## üß† Versi√≥ iterativa

````python
from typing import List

def fibonacci_iteratiu(n: int) -> int:
    """
    Calcula el n-√®ssim nombre de Fibonacci de forma iterativa.

    :param n: √çndex (n ‚â• 0) del nombre de Fibonacci a calcular.
    :return: El nombre de Fibonacci a la posici√≥ n.
    """
    if n == 0:
        return 0

    seq: List[int] = [0, 1]
    print(seq)
    for i in range(2, n + 1):
        seq.append(seq[i - 1] + seq[i - 2])
        print(seq)

    return seq[n]
````
