## Algorisme Rus


```python
def producte_rus(a: int, b: int) -> int:
    resultat = 0
    while a > 0:
        # Si a √©s senar, afegeix b al resultat
        if a % 2 == 1:
         resultat += b
        # Divideix a per 2 (divisi√≥ entera) i multiplica b per 2
        a //= 2
        b *= 2
    return resultat

# Multiplica dos enters a i b fent √∫s de l'algorisme del producte rus i retorna el resultat a*b
```
