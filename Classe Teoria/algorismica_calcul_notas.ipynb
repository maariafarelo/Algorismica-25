## üß† C√†lcul nota portfoli laboratori

```python
def calcul_portfoli_sobre_10(
    prob_temps: float,
    nota_prob1: float,
    nota_prob2: float
) -> float:
    return (
        min(5.0, prob_temps * 0.5) +
        nota_prob1 * 0.25 +
        nota_prob2 * 0.25
    )
```

```python
nota_portfoli_sobre_10:float = calcul_portfoli_sobre_10(8.0, 10.0, 7.0)
print(nota_portfoli_sobre_10)
```

### üß† Calcul global laboratori

```python
def calcul_global_lab_sobre_3(
    prova1: float,
    prova2: float,
    prova3: float,
    nota_portfoli: float
) -> float:
    return (
        prova1 * 0.05 +
        prova2 * 0.075 +
        prova3 * 0.075 +
        nota_portfoli * 0.1
    )
```

```python
nota_lab_sobre_3 = calcul_global_lab_sobre_3(9.0,7.0,3.5,8.25)
print(nota_lab_sobre_3)
print("sobre 10 seria: ",  nota_lab_sobre_3 / 3 * 10)
```

## üß† C√†lcul nota teoria

```python
def calcul_nota_examen_sobre_10(part_practica: float, part_teorica: float) -> float:
    if part_practica > 2.5 and part_teorica > 2.5:
        return part_practica + part_teorica
    else:
        return min(part_practica, part_teorica)
```

```python
def calcul_parcial(parcial1: float, parcial1_repetit: float) -> float:
    return parcial1_repetit
```

```python
def calcul_global_teoria_sobre_7(parcial: float, final: float) -> float:
    return parcial * 0.3 + final * 0.4
```

```python
def calcul_global_assignatura(nota_lab: float, nota_teoria: float) -> float:
    nota_lab_10 = nota_lab / 3 * 10
    nota_teoria_10 = nota_teoria / 7 * 10
    if nota_lab_10 >= 4.0 and nota_teoria_10 >= 4.0:
        return nota_lab + nota_teoria
    else:
        return min(4.0, nota_lab + nota_teoria)
```
