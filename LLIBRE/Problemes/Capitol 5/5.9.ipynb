### 5.9 Valor igual al seu índex

```python
def existeix(llista):
    """
    Comprova si existeix un element tal que llista[i] == i.
    
    Parameters
    ----------
    llista: list
    
    Returns
    -------
    bExisteix: bool
    """
    esquerra, dreta = 0, len(llista) - 1
    
    while esquerra <= dreta:
        mig = (esquerra + dreta) // 2
        if llista[mig] == mig:
            return True
        elif llista[mig] < mig:
            esquerra = mig + 1
        else:
            dreta = mig - 1
    
    return False
```

```python
assert existeix([1, 4, 5, 6, 7]) == False
assert existeix([0, 1, 2, 7]) == True
assert existeix([-1, 1, 4, 8]) == True
assert existeix([-5, -3, 0, 1, 5, 7]) == False
```
