### 5.4 El més petit que falta

```python
def el_mes_petit(llista):
    """
    Troba el primer enter que manca en la consecutivitat amb O(1) espai addicional.
    
    Parameters
    ----------
    llista: list
    
    Returns
    -------
    valor: int
        Segons les regles:
        - si hi ha forat → retorna el valor del forat
        - si el primer enter que manca és 0 → retorna 0
        - si no hi ha forat → retorna -1
    """
    n = len(llista)
    
    # Posicionar els números a la seva "posició correcta"
    for i in range(n):
        while 0 <= llista[i] < n and llista[i] != i and llista[i] != llista[llista[i]]:
            j = llista[i]
            llista[i], llista[j] = llista[j], llista[i]
    
    # Buscar el primer forat
    for i in range(n):
        if llista[i] != i:
            return i if i != 0 else 0
    
    # Si no hi ha forats
    return -1
```

```python
assert el_mes_petit([0, 1, 2, 3, 4, 7, 12]) == 5
assert el_mes_petit([1, 2, 3, 4, 7, 12]) == 0
assert el_mes_petit([0,1,2,3]) == -1
```

```python
#Complejitat O(1)
def el_mes_petit(llista):
    """
    Troba el primer enter que manca en la consecutivitat amb O(1) espai addicional.
    
    Parameters
    ----------
    llista: list
    
    Returns
    -------
    valor: int
        Segons les regles:
        - si hi ha forat → retorna el valor del forat
        - si el primer enter que manca és 0 → retorna 0
        - si no hi ha forat → retorna -1
    """
    n = len(llista)  # O(1) passos

    # Posicionar els números a la seva "posició correcta" (Cyclic Sort)
    for i in range(n):  # O(n) iteracions
        while 0 <= llista[i] < n and llista[i] != i and llista[i] != llista[llista[i]]:  # O(1) per moviment
            j = llista[i]  # O(1)
            llista[i], llista[j] = llista[j], llista[i]  # O(1) intercanvi
            # Cada element només es mou una vegada → total O(n) moviments

    # Buscar el primer forat
    for i in range(n):  # O(n) passos
        if llista[i] != i:  # O(1) comparació
            return i if i != 0 else 0  # O(1) retorn

    # Si no hi ha forats
    return -1  # O(1)
```
