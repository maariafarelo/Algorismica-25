### 5.8 Rotacions

```python
def rotacions(llista):
    """
    Troba el nombre de rotacions que s'han fet a una llista ordenada i rotada.
    
    Parameters
    ----------
    llista: list

    Returns
    -------
    numRotacions: int
    """
    esquerra, dreta = 0, len(llista) - 1
    
    # Si la llista no està rotada
    if llista[esquerra] <= llista[dreta]:
        return 0
    
    while esquerra <= dreta:
        mig = (esquerra + dreta) // 2
        # Comprovar si mig és el mínim
        if mig > 0 and llista[mig] < llista[mig - 1]:
            return mig
        elif llista[mig] >= llista[esquerra]:
            # La part esquerra està ordenada, anar a la dreta
            esquerra = mig + 1
        else:
            # La part dreta està ordenada, anar a l’esquerra
            dreta = mig - 1
    
    return 0  # per seguretat, tot i que no s’hauria d’arribar aquí
```

```python
assert rotacions([9, 10, 2, 5, 6, 8]) == 2
assert rotacions([3, 5, 10, 12, 1, 2]) == 4
assert rotacions([20, 2, 3, 7, 15, 18]) == 1
```

```python
def rotacions(llista):
    """
    Troba el nombre de rotacions que s'han fet a una llista ordenada i rotada.
    
    Parameters
    ----------
    llista: list

    Returns
    -------
    numRotacions: int
    """
    esquerra, dreta = 0, len(llista) - 1  # O(1) espai → només dues variables enters
    
    # Si la llista no està rotada
    if llista[esquerra] <= llista[dreta]:  # O(1) espai → comparació, no es crea cap estructura nova
        return 0                             # O(1) espai → retorn d’un enter

    while esquerra <= dreta:                 # O(1) espai → bucle no crea variables addicionals
        mig = (esquerra + dreta) // 2      # O(1) espai → variable temporal mig
        # Comprovar si mig és el mínim
        if mig > 0 and llista[mig] < llista[mig - 1]:  # O(1) espai → comparacions amb elements de llista
            return mig                               # O(1) espai → retorn
        elif llista[mig] >= llista[esquerra]:        # O(1) espai → comparació
            # La part esquerra està ordenada, anar a la dreta
            esquerra = mig + 1                        # O(1) espai → assignació
        else:
            # La part dreta està ordenada, anar a l’esquerra
            dreta = mig - 1                           # O(1) espai → assignació

    return 0  # O(1) espai → retorn
```
