### 5.5 Quantitat d'1s

```python
def quants_1(llistaBinaria):
    """
    Aquesta funció troba de manera eficient el nombre d'1s que conté una llista binària ordenada.
    
    Parameters
    ----------
    llistaBinaria: list
        la llista de nombres binaris ordenada (0s seguits de 1s)
        
    Returns
    -------
    num1s: int
        el nombre d'1s que conté
    """
    n = len(llistaBinaria)
    if n == 0:
        return 0

    # Cerca binària del primer 1
    esquerra, dreta = 0, n - 1
    primer_un = n  # per defecte si no hi ha 1s
    while esquerra <= dreta:
        mig = (esquerra + dreta) // 2
        if llistaBinaria[mig] == 1:
            primer_un = mig
            dreta = mig - 1
        else:
            esquerra = mig + 1

    # Nombre d'1s = total elements - índex del primer 1
    num1s = n - primer_un
    return num1s
```

```python
assert quants_1([0, 0, 1, 1, 1, 1, 1])  == 5
assert quants_1([0, 0, 0, 0, 1, 1])  == 2
assert quants_1([0, 0, 0, 0, 0, 0, 0])  == 0
assert quants_1([1, 1, 1])  == 3
```

```python
def quants_1(llistaBinaria):
    """
    Aquesta funció troba de manera eficient el nombre d'1s que conté una llista binària ordenada.
    
    Parameters
    ----------
    llistaBinaria: list
        la llista de nombres binaris ordenada (0s seguits de 1s)
        
    Returns
    -------
    num1s: int
        el nombre d'1s que conté
    """
    n = len(llistaBinaria)              # O(1) - comptar elements de la llista
    if n == 0:                          # O(1) - comparació
        return 0                        # O(1) - retorn

    # Inicialització de la cerca binària
    esquerra, dreta = 0, n - 1          # O(1)
    primer_un = n                        # O(1) - supòsit si no hi ha 1s

    # Cerca binària per trobar el primer 1
    while esquerra <= dreta:             # O(log n) iteracions
        mig = (esquerra + dreta) // 2   # O(1) - càlcul de l'índex mig
        if llistaBinaria[mig] == 1:     # O(1) - comparació
            primer_un = mig              # O(1) - assignació
            dreta = mig - 1              # O(1) - actualització límit dret
        else:
            esquerra = mig + 1           # O(1) - actualització límit esquerre

    # Càlcul del nombre d'1s
    num1s = n - primer_un                # O(1)
    return num1s                         # O(1)



#Complexitat O(log n)
```
