### 5.7 Zeros al final

```python
def zeros_final(llista):
    """
    Mou tots els zeros de la llista donada al final, in-place.
    És una variació de partition de quicksort utilitzant 0 com a pivot.
    
    Parameters
    ----------
    llista: list
    
    Returns
    -------
    None
    """
    n = len(llista)
    pivot_index = 0  # Índex on posarem el proper element no-zero

    for i in range(n):
        if llista[i] != 0:
            # Intercanviar element actual amb pivot_index
            llista[i], llista[pivot_index] = llista[pivot_index], llista[i]
            pivot_index += 1

```

```python
llista = [0,3,0,1,0,5,0,0,2,7,8] 
zeros_final(llista)
assert llista == [3,1,5,2,7,8,0,0,0,0,0]
```
