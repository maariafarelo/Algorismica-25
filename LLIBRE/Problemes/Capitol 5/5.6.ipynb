### 5.6 Negatius al davant

```python
def negatius(llista):
    """
    Posiciona els nombres negatius al davant ordenats de més petits a més grans,
    i els positius després, in-place.
    
    Parameters
    ----------
    llista: list
    
    Returns
    -------
    llistaOrdenada: list
    """
    llista = [1, -2, 3, -4, -3, 5, 6]

    # Separar els negatius i ordena'ls
    negatius = sorted([x for x in llista if x < 0])
    positius = [x for x in llista if x >= 0]
    
    # Reescriure la llista original in-place
    llista[:len(negatius)] = negatius
    llista[len(negatius):] = positius

    return llista
```
