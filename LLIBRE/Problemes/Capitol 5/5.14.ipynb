### 5.14 Sumatori equivalent a un valor amb llista ordenada

```python
def merge_sort(llista):
    if len(llista) <= 1:
        return llista
    
    mig = len(llista) // 2
    esquerra = merge_sort(llista[:mig])
    dreta = merge_sort(llista[mig:])
    
    return merge(esquerra, dreta)

def merge(esq, dre):
    i = j = 0
    resultat = []
    
    while i < len(esq) and j < len(dre):
        if esq[i] <= dre[j]:
            resultat.append(esq[i])
            i += 1
        else:
            resultat.append(dre[j])
            j += 1
    
    resultat.extend(esq[i:])
    resultat.extend(dre[j:])
    return resultat


def sumatori_parelles_ord(llista, valorSuma):
    """
    Troba totes les parelles de nombres a la llista que sumin el valor donat
    i retorna les parelles ordenades sense fer servir 'sorted'.
    """
    
    vist = set()
    parelles = set()   # Evitem duplicats
    
    for x in llista:
        y = valorSuma - x
        if y in vist:
            parella = tuple(sorted((x, y)))  # Solo ordenar la parella individual
            parelles.add(parella)
        vist.add(x)
    
    parelles_llista = list(parelles)
    
    # Ordenem alfabeticament per (a, b)
    parelles_ordenades = merge_sort(parelles_llista)
    
    return parelles_ordenades
```

```python
assert sumatori_parelles_ord([1, 2, 3, 5, 7, 8], 10) == [(2,8),(3,7)]
```

```python
def merge_sort(llista):
    if len(llista) <= 1:           # O(1)
        return llista              # O(1)
    
    mig = len(llista) // 2         # O(1)

    # crides recursives → O(n/2 log(n/2)) cadascuna
    esquerra = merge_sort(llista[:mig])   # O(n/2 log(n/2))
    dreta = merge_sort(llista[mig:])      # O(n/2 log(n/2))
    
    return merge(esquerra, dreta)         # O(n)


def merge(esq, dre):
    i = j = 0                      # O(1)
    resultat = []                  # O(1)
    
    while i < len(esq) and j < len(dre):     # O(n)
        if esq[i] <= dre[j]:                 # O(1)
            resultat.append(esq[i])           # O(1)
            i += 1                            # O(1)
        else:
            resultat.append(dre[j])           # O(1)
            j += 1                            # O(1)
    
    resultat.extend(esq[i:])       # O(n)
    resultat.extend(dre[j:])       # O(n)

    return resultat                # O(1)


def sumatori_parelles_ord(llista, valorSuma):
    """
    Troba totes les parelles de nombres a la llista que sumin el valor donat
    i retorna les parelles ordenades sense fer servir 'sorted'.
    """
    
    vist = set()                   # O(1)
    parelles = set()               # O(1)
    
    for x in llista:               # n iteracions → O(n)
        y = valorSuma - x          # O(1)

        if y in vist:              # O(1) amortitzat
            parella = tuple(sorted((x, y)))   # ordenar 2 elements → O(1)
            parelles.add(parella)             # O(1) amortitzat
        
        vist.add(x)                # O(1)
    
    parelles_llista = list(parelles)      # O(k)

    # Ordenem les parelles amb merge sort
    parelles_ordenades = merge_sort(parelles_llista)   # O(k log k)
    
    return parelles_ordenades       # O(1)


#total: O(n log n)
```
