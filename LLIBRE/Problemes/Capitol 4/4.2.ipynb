### 4.2 Alfabet aviació

```python
def aviacio(cadena):
    """
    Aquesta funció converteix una cadena d'entrada a l'alfabet fonètic.
    
    Parameters
    ----------
    cadena: string
    
    Returns
    -------
    traduccio: list
    """
    diccionari = {
        "A": "Apple",
        "B": "Banana",
        "Y": "Yankee",
        "T": "Tango",
        "H": "Hotel"
    }
    
    traduccio = []
    for lletra in cadena.upper():
        if lletra in diccionari:
            traduccio.append(diccionari[lletra])
        else:
            traduccio.append(None)  # o bé saltar-la
    
    return traduccio
```

```python
assert aviacio('YtH') == ['Yankee', 'Tango', 'Hotel']
```
