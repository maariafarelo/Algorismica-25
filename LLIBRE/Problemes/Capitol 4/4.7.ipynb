### 4.7 Subcadena més llarga sense cap caràcter repetit

<div style="background-color:rgba(0, 0, 0, 0.2); text-align:left; vertical-align: middle;  border-radius:10px">
    <div style="padding: 20px">
    <h3>Correcció d'errates: </h3>
        <p>La resolució es troba al final del llibre, sota l'etiqueta <B>"Problema 4.8, pàgina 60"</B>. Està mal etiquetat, hauria de ser <B>"Problema 4.7, pàgina 59"</B></p>
    </div>
</div>

```python
def subcadena_mes_llarga(cadena):
    """
    Aquesta funció identifica la subcadena més llarga sense cap caràcter repetit.
    """
    vista = {}              # caràcter -> última posició vista
    inici = 0               # inici de la finestra
    millor_inici = 0        # inici de la millor subcadena
    millor_longitud = 0     # longitud de la millor subcadena

    for i, c in enumerate(cadena):
        # Si el caràcter ja s'ha vist i és dins de la finestra actual
        if c in vista and vista[c] >= inici:
            # mou l'inici de la finestra
            inici = vista[c] + 1
        
        # actualitza la posició del caràcter
        vista[c] = i
        
        # comprova si és la millor subcadena fins ara
        llargada_actual = i - inici + 1
        if llargada_actual > millor_longitud:
            millor_longitud = llargada_actual
            millor_inici = inici

    return cadena[millor_inici : millor_inici + millor_longitud]
```

```python
assert subcadena_mes_llarga('lacadenamesllarga') == 'namesl'
assert subcadena_mes_llarga('mesllarga') == 'mesl'
assert subcadena_mes_llarga('aaa') ==  'a'
```

La complexitat de l'algorisme és O(N).
