### 2.12 Nombres amics

```python
def amics(num1, num2):
    """
    Aquesta funció comprova si dos nombres són amics.
    
    Parameters
    ----------
    num1: int
        El primer número
    num2: int
        El segon números
    
    Returns
    -------
    b: bool
        Si la parella de nombres d'entrada són amics    

    Dos nombres enters a i b es diuen amics si cadascun és igual 
    a la suma dels divisors propis de l’altre.
    """
    # Funció interna per calcular la suma de divisors propis
    def suma_divisors(n):
        suma = 0
        for i in range(1, n):
            if n % i == 0:
                suma += i
        return suma

    # Comprovem la condició de nombres amics
    b = (suma_divisors(num1) == num2) and (suma_divisors(num2) == num1)
    
    return b

```

```python
assert amics(220, 284) == True
assert amics(200, 230) == False
assert amics(9363584, 9437056) == True
```
