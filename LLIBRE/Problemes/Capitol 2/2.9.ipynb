### 2.9 Min max

```python
def minim_maxim(llista):
    """
    Aquesta funció troba el valor mínim i el valor màxim.
    
    Parameters
    ----------
    llista: list
        La llista d'enters
    
    Returns
    -------
    min, max: int, int
        Els valors mínim i màxim dins la llista
    """
    _min = llista[0]
    _max = llista[0]
    n = len(llista)

    for i in range(n):
        if llista[i] > _max:
            _max = llista[i]
        elif llista[i] < _min:
            _min = llista[i]

   
    return _min, _max

```

```python
assert minim_maxim([3, 1, 5, 2, 7, 8]) == (1, 8)
assert minim_maxim([1, 7, 4, 6, 8, -2, 9, 5]) == (-2, 9)
```
