### 2.11 Suma quadrats

```python
def suma_quadrats(n = 100):
    """
    Aquesta funció calcula la diferencia entre la suma dels quadrats i els quadrat de la suma.
    
    Parameters
    ----------
    n: int
    
    Returns
    -------
    diff: int
    """
    suma_quadrats = 0
    quadrat_suma = 0
    for i in range(n+1):
        suma_quadrats+= i*i
        quadrat_suma+= i
    quadrat_suma*= quadrat_suma

    s = quadrat_suma - suma_quadrats
    
    return s
```

```python
assert suma_quadrats() == 25164150
```
