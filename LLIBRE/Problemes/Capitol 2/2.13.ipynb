### 2.13  Nombres perfectes

```python
def perfecte(num):
    """
    Aquesta funció comprova si un nombre es perfecte.
    
    Parameters
    ----------
    num: int
        El número que es vol comprovar
    
    Returns
    -------
    b: bool
        Si el nombre d'entrada és un nombre perfecte
    és perfecte si és igual a la suma dels seus divisors
     propis (tots els divisors positius menys el mateix nombre).
    
    """
    
    suma = 0
    for i in range(1, num):
        if num % i == 0:
            suma += i
    
    if suma == num:
        b = True
    else:
        b = False
    
    return b
```

```python
assert perfecte(28) == True
assert perfecte(6) == True
assert perfecte(8) == False
```
