### 2.6 Palindrom

```python
import itertools

def generar_palindroms(cadena):
    """
    Aquesta funció crea tots els palíndroms possibles amb les lletres de la cadena donada.
    
    Parameters
    ----------
    cadena: string
        La cadena donada
    
    Returns
    -------
    None
    """  
    # Eliminem espais i passem a minúscules per coherència
    cadena = cadena.replace(" ", "").lower()
    
    # Si la cadena és massa curta, no hi ha palíndroms interessants
    if len(cadena) < 2:
        print("No hi ha palíndroms possibles.")
        return
    
    palindroms = set()
    
    # Generem totes les permutacions possibles de la cadena
    for perm in set(itertools.permutations(cadena)):
        text = ''.join(perm)
        if text == text[::-1]:  # comprovem si és palíndrom
            palindroms.add(text)
    
    if palindroms:
        print("Palíndroms possibles:")
        for p in sorted(palindroms):
            print(p)
    else:
        print("No s'han trobat palíndroms.")
```

```python
generar_palindroms('aarrj')
generar_palindroms('aann')
generar_palindroms('aabtb')
```
