```python
def es_capicua_rotacions(paraula):
    """
    Aquesta funció determina si una paraula és capicua aplicant rotacions.

    Parameters
    ---------
    paraula: string
        La paraula a determinar

    Returns
    -------
    b: bool
        True si alguna rotació de la paraula és capicua, False altrament
    """
    paraula = paraula.lower().replace(" ", "")
    n = len(paraula)

    # Si ja és capicua tal qual
    if paraula == paraula[::-1]:
        return True

    # Provar totes les rotacions possibles
    for i in range(1, n):
        rotada = paraula[i:] + paraula[:i]
        if rotada == rotada[::-1]:
            return True  # trobat!

    return False
```

```python
assert es_capicua_rotacions("BAABCC") == True
```
