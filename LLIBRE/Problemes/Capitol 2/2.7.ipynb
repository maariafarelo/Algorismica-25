### 2.7 Divisors

```python
def divisors(num):                 # 1 operación (definición de función, constante)
    list = []                      # 1 operación (asignación de lista vacía)
    for i in range(1, num + 1):    # (num) iteraciones → se repite num veces
        if num % i == 0:           # 1 operación por iteración (módulo)
            list.append(i)         # en promedio <= num veces, coste O(1) cada vez
    return list                    # 1 operación (retorno)


```

```python
assert divisors(8) == [1,2,4,8]
assert divisors(5) == [1,5]
assert divisors(28) == [1, 2, 4, 7, 14,28]
```
