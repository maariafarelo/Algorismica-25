### 2.16  Nombres ambiciosos

```python
def ambicios(num):
    """
    Aquesta funció comprova si un nombre és ambiciós.
    
    Parameters
    ----------
    num : int
        El número que es vol comprovar.
    
    Returns
    -------
    b : bool
        True si el nombre és ambiciós, False en cas contrari.
    """
    def suma_divisors(n):
        # Retorna la suma dels divisors propis de n (sense incloure n)
        return sum(i for i in range(1, n // 2 + 1) if n % i == 0)
    
    # Nombres perfectes coneguts (poden ampliar-se)
    perfectes = {6, 28, 496, 8128, 33550336}
    
    vist = set()
    actual = num
    
    while actual not in vist:
        vist.add(actual)
        actual = suma_divisors(actual)
        if actual in perfectes:
            return True
        if actual == 0:
            break
    return False
```

```python
assert ambicios(25) == True  
assert ambicios(100) == False 
assert ambicios(143) == True 
```

```python
def ambicios(num, vist=None):
    """
    Aquesta funció comprova recursivament si un nombre és ambiciós.
    
    Parameters
    ----------
    num : int
        El número que es vol comprovar.
    vist : set, opcional
        Conjunt dels nombres ja visitats (per evitar bucles).
    
    Returns
    -------
    bool
        True si el nombre és ambiciós, False en cas contrari.
    """
    
    def suma_divisors(n):
        # Suma dels divisors propis (sense incloure n)
        return sum(i for i in range(1, n // 2 + 1) if n % i == 0)
    
    perfectes = {6, 28, 496, 8128, 33550336}
    
    if vist is None:
        vist = set()
    
    # Cas base 1: el nombre és perfecte
    if num in perfectes:
        return True
    
    # Cas base 2: entrem en un bucle
    if num in vist or num == 0:
        return False
    
    # Marquem el nombre com visitat
    vist.add(num)
    
    # Crida recursiva amb la suma dels divisors
    return ambicios(suma_divisors(num), vist)
```
