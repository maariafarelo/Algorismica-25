### 2.10 Sumatori de Parelles

Una millor manera de dir-ho seria: “L’algorisme pot tenir una complexitat d’ordre $\mathcal{O}(n^2)$ o menys” ja que es pot resoldre amb complexitat menor que $n^2$.</p>

```python
def sumatori_parelles(llista, valorSuma):
    """
    Aquesta funció trobat totes les parelles de nombres que sumin el valor.
    
    Parameters
    ----------
    llista: list
        La llista d'enters
    valorSuma: int
        valor que 
    
    Returns
    -------
    parelles: list
        Una llista de parelles la suma de la qual és igual a valorSuma.
        Les parelles no s'han de repetir a la llista i el primer element de
        la parella serà el que tingui valor més baix.
    """
    parelles = []
    n = len(llista)
    for i in range (0,n):
        for j in range(i,n):
            if valorSuma == llista[i] + llista[j]:
                tupla = (llista[i], llista[j])
                parelles.append(tupla)
    parelles = set(parelles)
    return parelles
print(set(sumatori_parelles([3, 1, 5, 2, 7, 8], 10))

```

```python
assert set(sumatori_parelles([3, 1, 5, 2, 7, 8], 10))  == set([(2, 8), (3, 7)])
assert set(sumatori_parelles([4,12,7,6,1,3,2,9,11,5,8,20,13,0,14],20)) == set([(0, 20), (6, 14), (7, 13), (8, 12), (9, 11)])
```
