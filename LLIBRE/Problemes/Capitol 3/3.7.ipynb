```python
def calculadora():
    """
    Aquesta funció representa una calculadora per línia de comandes.
    Llegeix expressions aritmètiques fins que l'usuari escriu 'sortir'.
    
    Parameters
    ----------
    None
    
    Returns
    -------
    None
    """
    print("Calculadora senzilla. Escriu 'sortir' per acabar.")
    
    while True:
        entrada = input("Introdueix una operació (ex: 3 + 4): ")
        
        if entrada.lower() == "sortir":
            print("Fins aviat!")
            break
        
        try:
            # Calcula l'expressió
            resultat = eval(entrada)
            print("Resultat:", resultat)
        except Exception as e:
            print("Entrada no vàlida. Prova de nou.")
```

```python
def mes_petit_divisible_per(n):
    """
    Aquesta funció calcula el nombre més petit divisible per tots els nombres de 1 a n
    sense importar cap mòdul extern.
    
    Parameters
    ----------
    n: int
    
    Returns
    -------
    int
        Nombre més petit divisible per tots els enters de 1 a n
    """
    def gcd(a, b):
        """Màxim comú divisor amb l'algorisme d'Euclides"""
        while b != 0:
            a, b = b, a % b
        return a
    
    def lcm(a, b):
        """Mínim comú múltiple de dos nombres"""
        return a * b // gcd(a, b)
    
    resultat = 1
    for i in range(2, n + 1):  # Incloem n
        resultat = lcm(resultat, i)
    
    return resultat


```
