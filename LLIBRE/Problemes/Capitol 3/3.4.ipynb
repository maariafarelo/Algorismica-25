### 3.4 Resta Binària

```python
def resta_binaria(op1, op2):
    """
    Calcula la resta binaria de dos números representados como listas.
    Conserva la longitud original.
    """
    n = max(len(op1), len(op2))
    op1 = [0]*(n - len(op1)) + op1
    op2 = [0]*(n - len(op2)) + op2

    resultado = [0]*n
    prestamo = 0

    for i in range(n-1, -1, -1):
        resta = op1[i] - op2[i] - prestamo
        if resta < 0:
            resta += 2
            prestamo = 1
        else:
            prestamo = 0
        resultado[i] = resta

    return resultado
```

```python
assert resta_binaria([0,1,0,1], [0,0,0,1]) == [0,1,0,0]
```
