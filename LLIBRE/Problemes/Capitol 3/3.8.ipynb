```python
def eratostenes(n):
    """
    Torna tots els nombres primers fins a n sense usar break.
    
    Parameters
    ----------
    n: int
    
    Returns
    -------
    ret: list
        Llista amb tots els nombres primers fins a n
    """
    if n < 2:
        return []
    
    primers = []
    for num in range(2, n + 1):
        es_primer = True
        for i in range(2, int(num**0.5) + 1):
            es_primer = es_primer and (num % i != 0)
        if es_primer:
            primers.append(num)
    return primers
```
