        <p><B>1. Tres condicions</B> Fes una funció que trobi el nombre <I>n</I> més gran tal que compleixi les tres condicons següents: <br>&nbsp;&nbsp;&nbsp;&nbsp;1. n &lt; 10.000<br>&nbsp;&nbsp;&nbsp;&nbsp;2. n = m x m per algun m enter<br>&nbsp;&nbsp;&nbsp;&nbsp;3. n = k! + 1 per algun k.</p>

```python
def numero_3_condicions():
    """
    Troba el nombre més gran menor que 10000 que sigui quadrat perfecte i
    de la forma k! + 1.
    
    Returns
    -------
    buscat: int
    """
    buscat = 0
    factorial = 1
    k = 1
    
    while True:
        factorial *= k  # k!
        n = factorial + 1
        if n >= 10000:
            break
        # Comprovem si n és quadrat perfecte
        m = int(n ** 0.5)
        if m * m == n:
            buscat = n  # guardem el més gran que compleixi
        
        k += 1
    
    return buscat
```

```python
assert numero_3_condicions() == 5041
```

```python
def setmanes_segons():
    """
    Troba tots els nombres de setmanes N tals que el nombre de segons en N setmanes
    sigui un factorial K! amb K <= 15.
    
    Returns
    -------
    n_setmanes: list of int
        Llista amb tots els N possibles
    """
    n_setmanes = []
    segons_per_setmana = 7 * 24 * 60 * 60  # 604800 segons per setmana
    
    factorial = 1
    for k in range(1, 16):
        factorial *= k
        N = factorial / segons_per_setmana
        if N == int(N):  # només afegim si la divisió és exacta
            n_setmanes.append(int(N))
    
    return n_setmanes
```

```python
assert setmanes_segons() == [6, 66, 792, 10296, 144144]
```

```python
def quadrats(n):
    """
    Troba dos nombres positius (a, b) tals que a^2 + b^2 = n.
    Retorna (-1, -1) si no existeix cap solució.
    
    Parameters
    ----------
    n: int
    
    Returns
    -------
    (a, b): tuple of int
    """
    for a in range(1, int(n**0.5) + 1):
        b2 = n - a**2
        if b2 > 0:
            b = int(b2**0.5)
            if b * b == b2:
                return (a, b)
    return (-1, -1)
```

```python
assert quadrats(125) == (2,11)
assert quadrats(106) == (5,9)
assert quadrats(81) == (-1,-1)
```

```python
def multiples():
    """
    Retorna la suma de tots els naturals menors de 1000
    que són múltiples de 3 o 5.
    
    Returns
    -------
    suma: int
    """
    suma = 0
    for i in range(1, 1000):
        if i % 3 == 0 or i % 5 == 0:
            suma += i
    return suma

```

```python
assert multiples() == 233168
```
