### 3.2 El Màxim Comú Divisor i les seves aplicacions

```python
import math

def reduir_fraccio(numerador, denominador):
    """
    Funció que retorna l'expressió irreductible d'una fracció.
    
    Parameters
    ----------
    numerador : int
        El numerador de la fracció.
    denominador : int
        El denominador de la fracció.
    
    Returns
    -------
    numReduit : int
        Numerador reduït.
    denReduit : int
        Denominador reduït.
    """
    
    if denominador == 0:
        raise ValueError("El denominador no pot ser zero.")
    
    # Calcular el màxim comú divisor
    mcd = math.gcd(numerador, denominador)
    
    # Dividir numerador i denominador pel MCD
    numReduit = numerador // mcd
    denReduit = denominador // mcd
    
    # Normalitzar signe: el denominador sempre positiu
    if denReduit < 0:
        numReduit *= -1
        denReduit *= -1
    
    return (numReduit, denReduit)
```

```python
assert reduir_fraccio(12, 8) == (3,2)
```
